### My code below shows exponential growth. In particular, it shows that at 5% annual GDP growth, 100 years from now the society will be more than 100 times wealthier.

```python
import numpy as np
import matplotlib.pyplot as plt
r = 0.05
x = np.linspace(1,100,100)
y = 1*(1+r)**x
plt.plot(y)
plt.ylabel('GDP')
plt.xlabel('Time')
plt.show()
```

Implications:
- Economic growth compounds nonlinearly.
- Small difference in rate of growth has huge effect over long horizon.
